## Tema 4. Estructuras de decisión y repetición
Las decisiones se hacen con `if`, `elif`, `else`. Para repetir acciones se usa `for` con una lista o `range`.


```python
# Ejemplo: decisión + repetición
valores = [3, -1, 0, 7]

for v in valores:
    if v > 0:
        print(v, 'es positivo')
    elif v < 0:
        print(v, 'es negativo')
    else:
        print(v, 'es cero')
```
